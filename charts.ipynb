```python
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

df = pd.DataFrame({
    'Architecture': ['Single-Cloud', 'Hybrid-Cloud'],
    'Performance Efficiency': [2, 10],
    'Availability': [4, 7],
    'Learnaibility': [10, 10],
    'Operability': [10, 4],
    'Adaptability': [4, 8],
    'Scalability': [10, 8],
    'Cost': [2, 6]
})

categories=list(df)[1:]
N = len(categories)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

print(angles)

ax = plt.subplot(111, polar=True)

ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

plt.xticks(angles[:-1], categories)

ax.set_rlabel_position(0)
plt.yticks([2,4,6,8,10], ["2","4","6","8","10"], color="grey", size=7)
plt.ylim(0,10)

values=df.loc[0].drop('Architecture').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Single-Cloud")
ax.fill(angles, values, 'b', alpha=0.1)

values=df.loc[1].drop('Architecture').values.flatten().tolist()
values += values[:1]

ax.plot(angles, values, linewidth=1, linestyle='solid', label="Hybrid-Cloud")
ax.fill(angles, values, 'r', alpha=0.1)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

#plt.title("Architekturvergleich Single-Cloud vs. Hybrid-Cloud")

plt.show()
```
